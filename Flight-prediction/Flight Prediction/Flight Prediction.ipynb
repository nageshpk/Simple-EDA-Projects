# Flight Price prediction (EDA + Feature Engineering)

**Importing Libraries**

```python
import pandas as pd
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**Importing dataset**

```python
train_df = pd.read_excel('Data_train.xlsx')
train_df.head()
```

```python
test_df = pd.read_excel('Test_set.xlsx')
test_df.head()
```

```python
final_df = train_df.append(test_df)
final_df.head()
```

```python
final_df['Date_of_Journey'].str.split('/').str[1]
```

```python
# Feature engineering

final_df['Date'] = final_df['Date_of_Journey'].str.split('/').str[0]
final_df['Month'] = final_df['Date_of_Journey'].str.split('/').str[1]
final_df['Year'] = final_df['Date_of_Journey'].str.split('/').str[2]
```

```python
final_df.head()
```

```python
final_df.shape
```

```python
final_df['Date'] = final_df['Date'].astype(np.int64)
final_df['Month'] = final_df['Month'].astype(np.int64)
final_df['Year'] = final_df['Year'].astype(np.int64)
```

```python
final_df.info()
```

```python
final_df.drop('Date_of_Journey', axis=1, inplace=True)
```

```python
final_df.head()
```

```python
final_df['Arrival_Time'] = final_df['Arrival_Time'].str.split(' ').str[0]
```

```python
final_df['Arrival_hour'] = final_df['Arrival_Time'].str.split(':').str[0]
```

```python
final_df['Arrival_minutes'] = final_df['Arrival_Time'].str.split(':').str[1]
```

```python
final_df['Arrival_hour'] = final_df['Arrival_hour'].astype(np.int64)
final_df['Arrival_minutes'] = final_df['Arrival_minutes'].astype(np.int64)
```

```python
final_df.info()
```

```python
final_df.drop('Arrival_Time', axis = 1, inplace = True)
```

```python
final_df.head(1)
```

```python
final_df['Dep_hour'] = final_df['Dep_Time'].str.split(':').str[0]
final_df['Dep_minute'] = final_df['Dep_Time'].str.split(':').str[1]
final_df['Dep_hour'] = final_df['Dep_hour'].astype(np.int64)
final_df['Dep_minute'] = final_df['Dep_minute'].astype(np.int64)
final_df.drop('Dep_Time', axis = 1, inplace = True)
```

```python
final_df.info()
```

```python
final_df['Total_Stops'].unique()
```

```python
final_df[final_df['Total_Stops'].isnull()]
```

```python
final_df['Total_Stops'] = final_df['Total_Stops'].map({'non-stop': 0, '1 stop': 1, '2 stops': 2, '3 stops': 3, '4 stops': 4, 'nan': 1})
```

```python
final_df.drop('Route', axis = 1, inplace=True)
```

```python
final_df.head()
```

```python
final_df[final_df['Duration'] == '5m']
```

```python
final_df.drop(6474, axis = 0, inplace = True)
final_df.drop(2660, axis = 0, inplace = True)
```

```python
final_df['Duration_hour'] = final_df['Duration'].str.split(' ').str[0].str.split('h').str[0]
final_df['Duration_minute'] = final_df['Duration'].str.split(' ').str[1].str.split('m').str[0].fillna(0)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'].astype(np.int64)
final_df['Duration_minute'] = final_df['Duration_minute'].astype(np.int64)
```

```python
final_df['Duration_hour'] = final_df['Duration_hour'] * 60
```

```python
final_df.head()
```

```python
final_df['Duration'] = final_df['Duration_hour'] + final_df['Duration_minute']
```

```python
final_df.head()
```

```python
final_df.drop('Duration_hour', axis=1, inplace = True)
```

```python
final_df.drop('Duration_minute', axis = 1, inplace = True)
```

```python
final_df.head()
```

**Label encoding**

```python
final_df['Airline'].unique()
```

```python
from sklearn.preprocessing import LabelEncoder
labelencoder = LabelEncoder()
```

```python
final_df['Airline'] = labelencoder.fit_transform(final_df['Airline'])
final_df['Source'] = labelencoder.fit_transform(final_df['Source'])
final_df['Destination'] = labelencoder.fit_transform(final_df['Destination'])
final_df['Additional_Info'] = labelencoder.fit_transform(final_df['Additional_Info'])
```
